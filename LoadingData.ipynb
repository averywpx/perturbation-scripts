# Convert dataset to anndata

```python
def dataset_to_anndata(ds: datasets.arrow_dataset.Dataset, 
                       gene_dict: dict) -> anndata.AnnData:
    """
    Convert Dataset to AnnData format.
    
    Inputs:
        ds : datasets.arrow_dataset.Dataset
            Dataset containing 'gene_token_id', 'gene_expression', and 'cell_barcode' columns 
            plus metadata
        gene_dict : dict
            Dictionary mapping gene token IDs (int) to Ensembl gene names (str)
    
    Outputs:
        anndata.AnnData
            AnnData object with sparse expression matrix
    """
    
    # set-up gene mapping
    gene_token_ids = sorted(gene_dict.keys())
    token_to_col = {token: idx for idx, token in enumerate(gene_token_ids)}
    gene_names = [gene_dict[token] for token in gene_token_ids]
    
    # pre-allocate for CSR matrix components
    data, indices, ptr = [], [], [0]
    obs_records = []
    
    for i, row in enumerate(ds):
        genes = row['gene_token_id']
        expressions = row['gene_expression']
        
        # map to column indices
        gene_indices = [token_to_col[gene] for gene in genes]
        
        # add to CSR components
        data.extend(expressions)
        indices.extend(gene_indices)
        ptr.append(len(data))
        
        # extract metadata (exclude gene expression columns)
        metadata = {k: v for k, v in row.items() 
                   if k not in ['gene_token_id', 'gene_expression']}
        obs_records.append(metadata)
    
    # create sparse matrix
    n_cells = len(ptr) - 1
    n_genes = len(gene_names)

    csr_matrix_obj = csr_matrix((data, indices, ptr), shape=(n_cells, n_genes))

    # create AnnData
    obs_df = pd.DataFrame(obs_records)
    obs_df = obs_df.set_index('cell_barcode')  # set cell barcodes as index
    adata = anndata.AnnData(X=csr_matrix_obj, obs=obs_df)
    adata.var.index = pd.Index(gene_names, name='ensembl_id')
    
    return adata
```

```python
token_map = dict(zip(
    gene_metadata_df["gene_name"],
    gene_metadata_df["gene_token_id"]
))
token_map
```

```python
def knockdown(target_gene, adata, obs_col="gene_target", ntc="Non-Targeting"):
    
    idx = token_map[target_gene]

    
    gene_adata = adata[adata.obs[obs_col]==target_gene].copy()
    ntc_adata = adata[adata.obs[obs_col]==ntc].copy()

    gene_expression = gene_adata.X[:, idx].toarray().ravel()
    ntc_gene_expression = ntc_adata.X[:, idx].toarray().ravel()

    

    # pert = adata.obs[obs_col] == target_gene
    # ctrl = adata.obs[obs_col] == ntc
    # print(idx)
    # print(pert)

    pert_expr = gene_expression.mean()
    ctrl_expr = ntc_gene_expression.mean()

    log2FC = np.log2((pert_expr + 1e-6) / (ctrl_expr + 1e-6))
    knockdown = 1 - pert_expr / ctrl_expr
    
    print(target_gene)
    print(f"Mean target expression in perturbed: {pert_expr}")
    print(f"Mean target expression in control: {ctrl_expr}")
    print(f"log2FC: {log2FC}")
    print(f"knockdown: {knockdown}")
    print("\n")
    

    return {
        "gene": target_gene,
        "log2FC": log2FC,
        "knockdown": knockdown
    }
```

```python
def get_gene_index(adata, target_gene, gene_symbol_col=None):
    """
    Find the column index of target_gene in adata.X.
    Tries adata.var_names first, then adata.var[gene_symbol_col].
    """
    if target_gene in adata.var_names:
        return np.where(adata.var_names == target_gene)[0][0]

    if gene_symbol_col is not None and gene_symbol_col in adata.var.columns:
        matches = np.where(adata.var[gene_symbol_col].astype(str).values == target_gene)[0]
        if len(matches) > 0:
            return matches[0]

    return None


def get_vector(x):
    """
    Convert sparse or dense matrix slice to 1D numpy array.
    """
    if hasattr(x, "toarray"):
        return np.asarray(x.toarray()).ravel()
    return np.asarray(x).ravel()
```

```python
if target_col not in adata.obs.columns:
    raise ValueError(f"{target_col} not found in adata.obs. Available columns: {list(adata.obs.columns)}")

targets = sorted(adata.obs[target_col].dropna().unique())
targets = [x for x in targets if x != ntc_label]
targets
```

```python
results = []

for target in targets:
    gene_idx = get_gene_index(adata, target, gene_symbol_col=gene_symbol_col)

    if gene_idx is None:
        print(f"Skipping {target}: target gene not found in adata.var_names or {gene_symbol_col}")
        continue

    pert_mask = adata.obs[target_col] == target

    if pert_mask.sum() == 0:
        continue

    ntc_expr = get_vector(adata.X[ntc_mask, gene_idx])
    pert_expr = get_vector(adata.X[pert_mask, gene_idx])

    mean_ntc = np.mean(ntc_expr)
    mean_pert = np.mean(pert_expr)

    # Add small pseudocount for log2FC
    pseudocount = 1e-6
    log2fc = np.log2((mean_pert + pseudocount) / (mean_ntc + pseudocount))

    if mean_ntc > 0:
        knockdown_percent = 100 * (1 - mean_pert / mean_ntc)
    else:
        knockdown_percent = np.nan

    try:
        stat, pval = ranksums(pert_expr, ntc_expr)
    except Exception:
        stat, pval = np.nan, np.nan

    results.append({
        "target_gene": target,
        "n_perturbed_cells": int(pert_mask.sum()),
        "n_ntc_cells": int(ntc_mask.sum()),
        "mean_expression_ntc": mean_ntc,
        "mean_expression_perturbed": mean_pert,
        "log2FC_perturbed_vs_ntc": log2fc,
        "knockdown_percent": knockdown_percent,
        "wilcoxon_rank_sum_stat": stat,
        "wilcoxon_rank_sum_pvalue": pval,
        "gene_id_used": adata.var_names[gene_idx],
    })


# =========================
# Save results
# =========================

df = pd.DataFrame(results)
df = df.sort_values("knockdown_percent", ascending=False)

df.to_csv(output_csv, index=False)

print(df.head(20))
print(f"Saved result to: {output_csv}")
```
